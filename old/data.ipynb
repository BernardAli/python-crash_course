```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')

values = range(0, 6)
squares = [x ** 2 for x in values]
fig, ax = plt.subplots()
ax.plot(values, squares, linewidth=1)
ax.set_title("Square Numbers", fontsize=16)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
# Scatter Plots
values = range(1, 1001)
squares = [x ** 2 for x in values]
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(values, squares, s=20, c='red')
plt.show()
```

```python
values = range(1, 1001)
squares = [x ** 2 for x in values]
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(values, squares, s=20, c=squares, cmap=plt.cm.Blues)
plt.show()
```
